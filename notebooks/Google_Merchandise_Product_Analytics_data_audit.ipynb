<h3>metrics</h3>

```python
purchases = fact[fact["type"] == "purchase"]
```

```python
purchases["price_in_usd"].sum()
```

```python
len(purchases)
```

```python
round(purchases["price_in_usd"].mean(), 2)
```

```python
fact["event_date"] = fact["date"].dt.date
```

```python
dau = (
    fact.groupby("event_date")
    ["user_id"]
    .nunique()
    .reset_index(name="dau")
)
```

```python
dau.head()
```

```python
revenue_daily = (
    purchases
    .groupby("event_date")
    ["price_in_usd"]
    .sum()
    .reset_index(name="revenue")
)
```

```python
revenue_daily.head()
```

```python
event_summary = (
    fact.groupby("type")
    .agg(
        events=("user_id","count"),
        users=("user_id","nunique")
    )
    .reset_index()
)

event_summary
```

```python
category_revenue = (
    purchases
    .groupby("category")
    ["price_in_usd"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

category_revenue.head(15)
```
